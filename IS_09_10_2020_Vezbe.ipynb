```
# All imports that we gonna use
import pandas as pd
import matplotlib.pyplot as plt


# Read with pandas csv file
IPH = pd.read_csv('GDAS_2017_2019_IPH.csv', usecols=['Date', 'prss'])

# Remove NaN rows from list
IPH_df = pd.DataFrame(IPH).dropna()

IPH_Data = {
    'Prss': IPH_df.prss,
    'Date': IPH_df.Date
}

IPH_df = pd.DataFrame(IPH_Data, columns=['Prss', 'Date'])
IPH_df.plot(y='Prss', x='Date')
```
